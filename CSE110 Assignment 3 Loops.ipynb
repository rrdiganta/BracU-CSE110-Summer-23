```python
#Task 16
quantity=int(input())
sum=large=small=0
count=1
while count<=quantity:
  num=int(input())
  sum=sum+num
  if num>large:
    large=num
  elif num<small:
    small=num
  count+=1
print("Maximum", large)
print("Minimum", small)
print("Average is", sum/quantity)
```
