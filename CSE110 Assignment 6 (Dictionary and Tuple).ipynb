```python
#Task 8
a_dict={}
limit=int(input())
count=sum=0
while count<limit:
  key=str(input())
  value=int(input(key+":"))
  a_dict[key]=value
  sum+=value
  count+=1
avg=sum/limit
print("Average is", avg)
```

```python
#Task 11
string1=str(input())
string2=string1.upper()
result={}
for i in string2:
    if i not in result.keys():
        result[i]=1
    else:
        result[i]+=1
    if ' ' in result:
      del result[' ']
print(result)
```

```python
#Task 13
list1=[("a", 1), ("b", 2), ("a", 3), ("b", 1), ("a", 2), ("c", 1)]
result={}
for x in list1:
  if x[0] in result.keys():
    result[x[0]].append(x[1])
  else:
    result[x[0]]=[x[1]]
print(result)
```

```python
dict1 = {'a':59 , 'b':-82 , 'c':5 , 'd':-81 , 'e':53}
for i in dict1:
    j = 0
    k = 22
    while j < 5:
        if j % 2 == 0:
            k = dict1[i] + j - (8 + k % 6) / 3
            dict1[i] = dict1[i]+ int(k)
        else:
            k = dict1[i] + j - (6 - k % 8) * 3
            dict1[i] = dict1[i] - int(k)
        j += 1
    print(int(k))
    print(i + " -> " + str(dict1[i]))
```
