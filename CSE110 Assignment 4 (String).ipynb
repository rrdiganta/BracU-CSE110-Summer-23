```python
#Task 10
string=str(input())
a1=0
b1=0
str_len=len(string)
while string[a1]!=",":
  a1+=1
b2=a1+2

result=""
while b1<a1 and b2<str_len:
  result+=string[b1]+string[b2]
  b1+=1
  b2+=1

if b1<a1:
  result+=string[b1:a1]
else:
  result+=string[b2:str_len]

print(result)
```

```python
i=0
j=0
k=15
test = '<--cat'
while i < 5:
    k -= 1
    j = k
    while j > 10:
        if j % 2 == 0:
            test += '-->'
            test = test + str(i) + str(j // 2)
        else:
            test += '<--'
            test = test + str(i // 2) + str(j)
        print(test)
        j-=1
    i+=1
```

```python
line = input()
n1 = 0
i1 = 0
while line[n1] != ',':
    n1 += 1
i2 = n1 + 2
n2 = len(line)

res = ''
while i1 < n1 and i2 < n2:
    res += line[i1] + line[i2]
    i1 += 1
    i2 += 1

if i1 < n1:
    res += line[i1:n1]
else:
    res += line[i2:n2]

print(res)
```
