```python
#Task 2
def fibonacci(a):
  num1=0
  num2=1
  print(num1, end=" ")
  print(num2, end=" ")
  for i in range(a):
    sum=num1+num2
    if sum<=a:
      print(sum, end=" ")
    num1=num2
    num2=sum
fibonacci(5)
```

```python
#Task 4
def funtion_name(string):
  count_upper=count_lower=0
  for i in string:
    if ord(i)>=65 and ord(i)<=90:
      count_upper+=1
    elif ord(i)>=97 and ord(i)<=122:
      count_lower+=1
  print("No. of Uppercase Characters:", count_upper)
  print("No. of Lowercase Characters:", count_lower)
funtion_name("HaRRy PotteR")
```

```python
#Task 7
def show_palindrome(num):
  result=""
  for m in range(1, num+1):
    result+=str(m)
  for n in range(num-1,0,-1):
    result+=str(n)
  return result
x=show_palindrome(5)
print(x)
```

```python
#Task 8 (ok)
def show_palindromic_triangle(num):
    for row in range(1, num+1):
        palindrome=show_palindrome(row)
        print(" "*(num-row)+palindrome+" ")
show_palindromic_triangle(4)
```

```python
#Task 8
def show_palindromic_triangle(num):
    for row in range(1, num+1):
        palindrome=show_palindrome(row)
        string=" ".join(palindrome)
        print(string.center(100))
show_palindromic_triangle(4)
```

```python
#Task 9
def area_circumference_generator(R):
  import math
  area=math.pi*R**2
  circumference=2*math.pi*R
  return (area, circumference)
tup=area_circumference_generator(1.5)
print(tup)
(x, y)=tup
print("Area of the circle is", x, "and circumference is", y)
```

```python
#Task 11
def rem_duplicate(tup):
  list1=[]
  for x in tup:
    if x not in list1:
      list1.append(x)
  return tuple(list1)
print(rem_duplicate((1,1,1,2,3,4,5,6,6,6,6,4,0,0,0)))
```

```python
#Task 13
def basic_calc(sign, x, y):
  if sign=="+":
    result=x+y
  elif sign=="-":
    result=x-y
  elif sign=="*":
    result=x*y
  elif sign=="/":
    if y==0:
      result="No Calculation Possible"
    else:
      result=x/y
  else:
    result="No Calculation Possible"
  return result
operator=str(input())
num1=int(input())
num2=int(input())
print(basic_calc(operator, num1, num2))
```

```python
#Task 14
def removal(string, num):
  result1=string[0]
  result2=""
  for i in range(1, len(string)):
    if i%num!=0:
      result1+=string[i]
    else:
      result2+=string[i]
  return result1+result2
sentence=str(input())
position=int(input())
print(removal(sentence, position))
```

```python
#Task 15
def pay_amount(groceries, location="Dhanmondi"):
  price_list={"Rice":105, "Potato":20, "Chicken":250, "Beef":510, "Oil":85}
  fee=0
  for i in groceries:
    if i in price_list:
      fee+=price_list[i]
  if location!="Dhanmondi":
    fee+=70
  else:
    fee+=30
  return fee
x=str(input())
order_items=x.replace("[", "").replace("]", "").replace('"', '').split(", ")
print(pay_amount(order_items))
```

```python
#Task 16
def splitting_money(money):
    fhnotes=str(money//500)
    ohnotes=str((money%500)//100)
    twnotes=str((money%100)//20)
    tenotes=str((money%20)//10)
    twonotes=str((money%10)//2)
    one_notes=str((money%2)//1)
    result= "500 Taka: "+fhnotes+" note(s)\n"+\
    "100 Taka: "+ohnotes+" note(s)\n"+\
    "20 Taka: "+twnotes+" note(s)\n"+\
    "10 Taka: "+tenotes+" note(s)\n"+\
    "2 Taka: "+twonotes+" note(s)\n"+\
    "1 Taka: "+one_notes+" note(s)"
    return result
print(splitting_money(1234))
```

```python
#Task 18
def divisible(start, end, fdiv, sdiv):
  sum=0
  for num in range(start, end):
    if (num%fdiv==0 or num%sdiv==0) and (num%fdiv!=0 or num%sdiv!=0):
      sum+=num
  return sum
START=int(input())
END=int(input())
FDIV=int(input())
SDIV=int(input())
print(divisible(START, END, FDIV, SDIV))
```
