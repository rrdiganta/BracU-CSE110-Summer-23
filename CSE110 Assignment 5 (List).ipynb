```python
#Task 16
numbers=str(input())
no_com=numbers.split(",")
print(no_com)
result=[]
i=max1=max2=0
for x in no_com:
  result.append(int(x))
print("My List:", result)
for y in result:
  if y>max1:
    max2=max1
    max1=y
  elif y>max2 and max1!=y:
    max2=y
  elif max2==max1 and max2!=y:
    max2=y
i=result.index(max2)
print("Second largest number in the list is", max2, "which was found at index", i)
```

```python
#Task 17
numbers=str(input())
no_com=numbers.split(",")
result=[]
m=n=max=min=0
for x in no_com:
  result.append(int(x))
print("My List:", result)
for y in range(len(result)):
  if result[y]>max:
    max=result[y]
    m=y
for y in result:
  if y!=0:
    if y<min:
      min=y
      n=result.index(y)
print("Smallest number in the list is", min, "which was found at index", n)
print("Largest number in the list is", max, "which was found at index", m)
```
